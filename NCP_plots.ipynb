```python
def plot_min_envelope(phi_by_k_dict, name):
    k_list = list(phi_by_k_dict.keys())
    min_pts = [min(phi_by_k_dict[k]) for k in k_list]
    max_k_to_plot = int(len(k_list)/2)+1
    plt.plot(k_list[:max_k_to_plot], min_pts[:max_k_to_plot],'-o', color='red')

    ax = plt.gca()
    ax.set_xlabel('k')
    ax.set_ylabel('Conductance')
    ax.set_yscale('log')
    ax.set_xscale('log')

    # Add-ons specific to different graphs
    if name=='karate_club':
        plt.axvline(x = 5, color='blue', ls='--', label = 'k=5')
        plt.axvline(x = 17, color='green',ls='--', label = 'k=17')
        ax.set_ylim(bottom=0.1, top=1)
        plt.title('Karate club')
        filename = 'karate_club_hypergraph_NCP.pdf'
    elif name=='dolphins':
        plt.axvline(x = 21, color='blue', ls='--', label = 'k=21')
        ax.set_ylim(bottom=0.01,top=1)
        ax.set_xlim(right=100)
        plt.title('Dolphins')
        filename = 'dolphins_hypergraph_NCP.pdf'

    ax.set_aspect('equal')
    plt.legend()
    plt.savefig('./NCP_figures/'+filename, bbox_inches='tight')
    plt.show()
    return
```

```python
phi_by_k, best_cut_by_k = sample_cuts_from_all_nodes(hgraph)
```

```python
plot_min_envelope(phi_by_k, name='karate_club')
```
